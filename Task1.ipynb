Univariate Analysis

```python
df_clean.duplicated
```

```python
sns.countplot(x='type', data=df_clean)
plt.title('Distribution of Content Type (Movie vs TV Show)')
```

```python
df_clean['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10, 5))
plt.title("Number of Titles Released Each Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='rating', hue='type', order=df['rating'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Type vs Rating")
plt.show()
```

```python
sns.boxplot(df_clean["rating"])
ylabel='rating'
```

```python
sns.scatterplot(x=df_clean["country"] , y =df_clean["rating"])
plt.show()
```
